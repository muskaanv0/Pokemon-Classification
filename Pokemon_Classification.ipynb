```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pokemon = pd.read_csv("pokemon.csv")
```

```python
pokemon.head()
```

```python
pokemon.shape
```

```python
y = pokemon['target'].values

X = pokemon.drop(columns=['target'], inplace=False).values
```

```python
X.shape
```

```python
y.shape
```

```python
plt.imshow(X[100].reshape(100,100,3))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
print( X_train.shape, y_train.shape )
print( X_test.shape, y_test.shape)
```

# **KNN** **METHOD**

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=13)
```

```python
# model training
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
# accuracy for the test set
model.score(X_test, y_test)
```

```python
# Accuracy achieved for k=13 is 0.7894736842105263
```

```python
#Predicting for any random Image
random_img = X_test[30]
plt.imshow(random_img.reshape(100,100,3))
```

```python
# actual output
y_test[30]
```

```python
# model prediction
model.predict( [random_img] )
```

```python
# We can see both actual output and the predication is the same
```

```python
for i in [11, 22, 23, 39, 55, 60]:
    random_img = X_test[i]
    plt.imshow(random_img.reshape(100,100,3))
    plt.show()
    print("Actual Label : ",y_test[i])
    p = model.predict( [random_img] )
    print("Predicted Label : ", p)
```

```python
# for both i=39 and i=55 this model has made wrong prediction.
```

# **LOGISTIC REGRESSION METHOD**


```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
# accuracy for the test set
model.score(X_test, y_test)
```

# **DECISION TREE METHOD**

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
```
